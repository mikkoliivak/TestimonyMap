```python
m = folium.Map(
    location=[42.9, -77.8],
    zoom_start=8,
    tiles='OpenStreetMap'
)

for dc in datacenter_data:
    num_testimonies = len(dc['testimonies'])
    
    if num_testimonies == 0:
        color = 'gray'
    elif num_testimonies < 5:
        color = 'orange'
    else:
        color = 'red'
    
    popup_html = f"""
    <div style='max-width: 450px; max-height: 600px; overflow-y: auto; font-family: Arial, sans-serif;'>
        <h3 style='margin-top: 0; color: #2c3e50; border-bottom: 2px solid #3498db; padding-bottom: 5px;'>
            {dc['name']}
        </h3>
        <p style='color: #7f8c8d; font-size: 13px;'>
            <strong>County:</strong> {dc['county']}<br>
            <strong>Total Testimonies:</strong> {num_testimonies}
        </p>
    """
    
    if dc['testimonies']:
        popup_html += "<hr style='border: 1px solid #ecf0f1;'>"
        
        for i, testimony in enumerate(dc['testimonies'], 1):
            popup_html += f"""
            <div style='margin-bottom: 20px; padding: 12px; background: #f8f9fa; border-left: 3px solid #3498db; border-radius: 3px;'>
                <h4 style='margin: 0 0 10px 0; color: #2c3e50; font-size: 14px;'>
                    Testimony {i} of {num_testimonies}
                </h4>
                <p style='font-style: italic; line-height: 1.6; color: #34495e; margin: 10px 0;'>
                    "{testimony['statement']}"
                </p>
                <div style='font-size: 12px; color: #7f8c8d; margin-top: 10px;'>
                    <p style='margin: 3px 0;'><strong>Date:</strong> {testimony['date']}</p>
                    <p style='margin: 3px 0;'><strong>Source:</strong> {testimony['source-details']}</p>
                    <p style='margin: 8px 0 0 0;'>
                        <a href='{testimony['source']}' target='_blank' style='color: #3498db; text-decoration: none; font-weight: bold;'>
                            📄 Read Full Article →
                        </a>
                    </p>
                </div>
            </div>
            """
    else:
        popup_html += """
        <p style='color: #95a5a6; font-style: italic;'>
            No testimonies available for this location.
        </p>
        """
    
    popup_html += "</div>"
    
    folium.Marker(
        location=[dc['lat'], dc['lng']],
        popup=folium.Popup(popup_html, max_width=500),
        tooltip=f"{dc['name']} - Click for testimonies",
        icon=folium.Icon(
            color=color,
            icon='server',
            prefix='fa'
        )
    ).add_to(m)

legend_html = """
<div style='position: fixed; 
            bottom: 50px; 
            left: 50px; 
            width: 200px; 
            background-color: white; 
            border: 2px solid grey; 
            border-radius: 5px; 
            z-index: 9999; 
            font-size: 14px;
            padding: 10px;'>
    <h4 style='margin-top: 0;'>Legend</h4>
    <p><span style='color: red;'>🔴</span> 5+ testimonies</p>
    <p><span style='color: orange;'>🟠</span> 1-4 testimonies</p>
    <p><span style='color: gray;'>⚪</span> No testimonies</p>
</div>
"""
m.get_root().html.add_child(folium.Element(legend_html))

m
```
